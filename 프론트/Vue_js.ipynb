Vue.js는 웹프론트엔드 프레임 워크

- 컴포넌트 기반의 SPA(Single Page Application)를 구축할 수 있게 해주는 프레임워크

컴포넌트
- 웹을 구성하는 로고, 메뉴바, 버튼, 모달창 등의 웹 페이지 내의 다양한 UI요소
- 재사용 가능하도록 구조화 한것

SPA(Single Page Application)
- 단일 페이지 어플리케이션
- 하나의 페이지안에서 필요한 영역부분만 로딩되는 형태
- 빠른 페이지 변환
- 적은 트래픽 양

단점 : 최초 로딩 시 시간이 오래 걸림(파일사이즈가 크면 클수록)


라우터(Router)
- Vue에서 Routing기능을 사용할 수 있게 해주는 공식 라이브러리
- 라우팅이라는 것은 웹페이지 간에 이동하는 방법을 말함
- 메뉴를 클릭하거나 페이지를 이동 할 때 새로운 페이지를 받아오는 형식을 하는데 Router는 SPA로 미리 해당하는 모든 컴포넌트 페이지를 받고, 라우팅을 사용해
  그 부분만 갱신 하는것(Vue, Angular도 모두 라우팅을 사용하여, 페이지를 이동함)

컴포넌트(Component)
- template
  - 탬플릿은 HTML과 vue에서 제공하는 디렉티브, 탬플릿, 다른 컴포넌트등이 들어가는 자리이다
  - 간단하게 HTML이 들어간다고 생각하면 편하다. 주의할 점으로는 하나의 element가 젱리 밖에 존재해야 한다

- script
  - JS가 들어가서 동작을 정의하는 부분이다.

- style
  - 말그대로 CSS나 SCSS같은 스타일이 들어가는 곳이다.

- Data
  - vue component에서는 무언가를 return하는 함수를 넣어줘야한다.
  - 만약 Vue객체의 인스턴스를 선언 한다면, 함수 대신 객체를 넣어도 괜찮다.

  <script>
  export default{
    data : function(){
      return{
        msg:"Hello World"
      }
    }
  }
  </script>
  
  - 이렇게 되면 vue-if, vue-model,{{}}을 쓸 수 있다.

- method
  - <template>
      <div>
        <button v-on:click="toggle">클릭!</button>
        <div v-if="show">이건 보이지않아요</div>
      </div>
    <template>
    <script>
      export default{
        data : function(){
          return{
            show:false
          }
        }
      },
      methods:{
        toggle:function(){
          this.show != this.show
        }
      }
    </script>

  - method 객체에는 함수들이 들어가며 이 함수들은 다른 method에서 혹은 탬플릿에서 사용될 수 있다
  - method안에도 화살표 함수를 사용하면 된다. this를 인식하지 못하는 사태가 벌어진다.

- computed
  - computed는 method와 유사하지만 computed에서 사용하고 있는 data에 변화가 있으면 자동으로 계산되는 함수들의 모음이다.
  <template>
    <div>
      {{value}}
    </div>
    <div>
      {{doubleValue}}
    </div>
    <button v-onclick="plus">클릭!</button>
  </template>
  <script>
    export default{
      name : "helloWorld",
      data(){
        return {
          value : 10
        }
      },
      methods:{
        plus:function(){
          this.value++
        }
      },
      computed:{
        doubleValue:function(){
          return this.value *2
        }
      }
    }
  </script>
  - 클릭을 할 때 마다, value가 1씩 증가하고 value가 바뀌기 때문에 doubleValue에 의해 아래 div가 2씩 증가하게 된다
  - computed도 마찬가지로 화살표 함수를 사용하면 안된다
  - computed는 methods와 달리 캐싱이라는 개념이 있어, 같은 페이지 내에서 같은 연산을 여러번 해야 한다면
  성능면에서 효율적으로 사용할 수 있다

- watch
  - watch는 computed와 비슷하지만 많은 옵션이 존재한다. 데이터 변경의 응답으로 비동기식 계산이 필요한 경우
  나 시간이 많이 소요되는 계산을 할 때 watch를 사용하는 편이 좋다.

Vue의 script
- 스크립트에서는 다른 모듈을 import할 수 있고, script내에서 export를 통해 컴포넌트를 조절하거나, 컴포넌트에서 필요한
data들을 제공하기도 한다.
