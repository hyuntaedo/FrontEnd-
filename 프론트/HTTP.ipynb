## Get & Post
- Get요청 : 데이터를 URL 내에 포함하여 요청할 수 있음(주로 리소스 요청에 사용)
- Post요청 : 데이터를 Form data에 퐇마하여 전달(주로 로그인에 사용)


- 웹 사이트를 생성하기 위한 언어로 문서와 문서가 링크로 연결되어 있고 태그를 사용하는 언어

###Get 요청하기
- http get 요청하기
- query parameter 이용하여 데이터 전달하기

```python
from bs4 import BeautifulSoup
import requests
url = 'https://sports.news.naver.com/news?oid=311&aid=0001321315'
```

```python
req = requests.get(url)
```

```python
# 200이면 정상인것
print(req)
req.text
```

### Post 요청하기
- http post 요청하기
- post data 이용하여 데이터 전달하기


```python
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id':'macamath22',
    'pwd':'Test1357!'
}
req = requests.post(url,data=data)
req.text
```

### HTTP header 데이터 이용하기
- header 데이터 구성하기
- header 데이터 전달하기


```python
url = 'https://sports.news.naver.com/news?oid=477&aid=0000308953'
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
resp = requests.get(url,headers=headers)
resp.text
```

### HTTP response 처리하기
- response 객체의 이해
- state_code 확인하기
- text 속성 확인하기

```python
url = 'https://sports.news.naver.com/news?oid=477&aid=0000308953'
resp = requests.get(url)
if resp.status_code == 200:
    resp.headers
else:
    print('error')
```
