# **MVC & MTV**

- Model : 안전하게 데이터를 저장
- View : 데이터를 적절하게 유저에게 보여줌
- Control, Template(Django) : 사용자의 입력과 이벤트에 반응하여 Model과 Veiw를 업데이트


- 예전에는 디자인, 로직, 데이터베이스가 한 곳에 뭉쳐있어서, 협업할 때 불편한 점이 많았음(디자이너가 control관련된 코드를 몰라서, 건드렸을 때 전체 시스템에 큰 문제가 발생하는 경우가 종종 있었음)
- 도저히 이렇게는 프로젝트 관리가 안되서 MVC모델을 적용해서 코드를 분리하게 된 것이다.
- 거의 대부분의 프로젝트들은 MVC(Model,Control,View)로 분리를 하면, 시스템 구성이 가능하다는 결론이 났음

- 액션이 발생하면 Django 서버로 들어온다.
- URL분석
- 분석한 URL에 적합한 View로 보내게됨
- 데이터베이스 어디에 접근해서 어떠한 데이터를 가공해야 할 지 알려주면
- 모델에서 실제로 데이터베이스와 커넥션을 하여 데이터를 가져오게 된다.
- 다시 그럼 데이터베이스에서 모델로 보내고 View로 보내고 View가 실질적으로 보여주는 데이터를 Tempalte로 전송하게됨
- Template가 HTML, Javascript등으로 연결되어 web으로 보내주게 된다.

- 녹색으로 되어있는 부분이 django가 다루게 될 것임


- WSGI(게이트웨이) : 웹서버와 Django를 결합하는 역할을 한다.
- URL Resolution : 파이썬 코드를 받게 되는 역할임(정규표현식으로 구성되어있어 정규표현식에 알맞은 View를 보여주게 된다.)
- VIEW : 사용자의 요청에의해 데이터를 입력을 하든지, 가지고와서 리스트를 보여줄 것인지 결정하게 된다.
- Model : 데이터 베이스로 부터 데이터를 가지고 올 때, 데이터를 다시 클래스에 넣고 하는 과정을 통합시킴(Database <-> Manager <-> MODEL)
  - 쿼리문을 사용할 필요가 없음
- Template : HTML파일 안에 로직을 삽입할 수 있음(컨트롤 로직)
  - View에서 받은 파일을 HTML로 보여 주는 파일
- FORM : 사용자 UI, 모델, 이런것들이 쉽게 관리가 된다.

# **Settings.py**

- 프로젝트 환경설정 파일
  - DEBUG
    - 디버그 모드 설정
  - INSTALLED_APPS
    - pip으로 설치한 앱 또는 본인이 만든 app을 추가
  - MIDDLEWARE_CLASSES
    - request와 response 사이의 주요 기능 레이어
  - TEMPLATES
    - django template관련설정, 실제 view(html, varible)
  - DATABASE
    - 데이터베이스 엔진의 연결 설정
  - STATIC_URL
    - 정적파일의 URL(css, javascript, image, etc...)


# **Manage.py**

- 프로젝트 관리 명령어 모음
  - 주요 명령어
    - startapp - 앱서버 생성
    - runserver - 서버실행
    - createsuperuser - 관리자 생성
    - makemigration app - app의 모델 변경사항 체크
    - migrate - 변경사항을 DB에 반영
    - shell - 쉘을 통해 데이터를 확인
    - collectstatic - static파일을 한 곳에 모음
- ex)
  - ./manage.py runserver 0.0.0.0:8080
